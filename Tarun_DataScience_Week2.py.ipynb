**Task 1: Basic Arithmetic Operations**

```python
# Function to perform and display arithmetic operations
def basic_operations(a, b):
    
    # Addition
    print(f"Addition: {a + b}")  
    
    # Subtraction
    print(f"Subtraction: {a - b}")  
    
    # Multiplication
    print(f"Multiplication: {a * b}")  
    
    # Division (check to avoid division by zero)
    print(f"Division: {a / b}" if b != 0 else "Division: Undefined")  
    
    # Modulus (remainder) with zero check
    print(f"Modulus: {a % b}" if b != 0 else "Modulus: Undefined")  
    
    # Exponentiation (a raised to the power b)
    print(f"Exponentiation: {a ** b}")  

# Taking user input
a = int(input("Enter the first integer: "))   
b = int(input("Enter the second integer: ")) 

# Calling the function to display results
basic_operations(a, b)
```

**Task 2: : Working with Lists and Arrays**

```python
import numpy as np

# Creating a list of at least 10 numbers
numbers = [10, 4, 23, 7, 56, 12, 89, 3, 15, 42]

print("Original List:", numbers)

# (a) Print length
print("Length of List:", len(numbers))

# (b) Max & Min
print("Maximum Value:", max(numbers))
print("Minimum Value:", min(numbers))

# (c) Add and Remove an element
numbers.append(100)       # adding
numbers.remove(7)         # removing specific value

print("List after adding 100 and removing 7:", numbers)

# (d) Sorting
asc = sorted(numbers)
desc = sorted(numbers, reverse=True)

print("Ascending Order:", asc)
print("Descending Order:", desc)

# Convert to NumPy array
arr = np.array(numbers)

# Mean, Median, SD
print("- NumPy Calculations -")
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

**Task 3: Dictionaries and Sets**

```python
# 1. Create a dictionary
student = {
    "name": "Tarun",
    "age": 22,
    "course": "Data Analytics",
    "marks": 88
}

# 2. Print each key-value pair
print("- Student Details -")
for key, value in student.items():
    print(f"{key}: {value}")

# 3. Add a new key
student["grade"] = "A"
print("After adding grade:", student)

# 4. Unique courses set
courses = {"Python", "Data Science", "AI", "Python"}
print("Unique Courses Set:", courses)

# 5. Set operations
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7}

print("- Set Operations -")
print("Union:", set_a.union(set_b))
print("Intersection:", set_a.intersection(set_b))
print("Difference:", set_a.difference(set_b))
```

**Task 4: File Handling**

```python
# 1. Student data (name, course, marks)
students = [
    ("Tarun", "Data Analytics", 88),
    ("Riya", "Python", 72),
    ("Aman", "AI", 95),
    ("Neha", "Data Science", 67),
    ("Kiran", "Machine Learning", 81)
]

# 2. Write data to student_data.txt
with open("student_data.txt", "w") as file:
    for name, course, marks in students:
        file.write(f"{name},{course},{marks}\n")

# 3. Read and display students with marks > 75
with open("student_data.txt", "r") as file:
    print("Students with Marks Above 75:")
    for line in file:
        name, course, marks = line.strip().split(",")
        marks = int(marks)

        if marks > 75:
            print(name, course, marks)
```
